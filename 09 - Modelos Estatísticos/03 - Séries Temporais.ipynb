#### Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
from statsmodels.tsa.api import ExponentialSmoothing
from statsmodels.tsa.api import SimpleExpSmoothing
from statsmodels.tsa.api import Holt
import warnings

warnings.filterwarnings('ignore')
```

#### Função para configurar os plots da Matplotlib

```python
def configurar_plot_com_dimensoes(titulo, x, y, h, w):
    plt.title(titulo)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.gcf().set_size_inches(h, w)
    plt.show()
```

```python
def configurar_plot(titulo, x, y):
    configurar_plot_com_dimensoes(titulo, x, y, 16, 8)
```

#### Importando e transformando os dados

```python
def definir_dados():
    dados = pd.read_excel('dados/Dados de Emprego_IA.xlsx')

    dados = dados.rename(columns = {
        'Mês': 'data',
        'Comercio': 'comercio',
        'Alimento': 'alimento',
        'Metalurgia': 'metalurgia'
    })
    
    return dados

dados = definir_dados()
    
dados.head()
```

#### Análise exploratória dos dados - Visualizando as 3 categorias pela data

```python
def analisar_item_por_data(item):
    dados_agrupados = dados[['data', item]].groupby('data', as_index = True).sum()
    dados_agrupados.plot(kind = 'line')
    configurar_plot(
        'Análise de aumento dos valores de {} de 2003 a 2007'.format(item),
        item, 
        'Data'
    )
```

```python
analisar_item_por_data('comercio')
analisar_item_por_data('alimento')
analisar_item_por_data('metalurgia')
```

Podemos observar, com base nas 3 análises, que podemos aplicar os seguintes algoritmos para cada um:

* Comércio - Modelo Holt Winters
* Alimento - Modelo Holt Winters
* Metalurgia - Média Movel Simples (MMS)

```python
dados_agrupados = dados[['data', 'metalurgia']].groupby('data', as_index = True).sum()
```

#### Aplicando o algoritmo de Média Movel SImples (MMS) nos dados de metalurgia

```python
dados_agrupados['media_movel_simples'] = dados_agrupados['metalurgia'].rolling(window = 3).mean()

dados_agrupados.plot(kind = 'line')
configurar_plot(
    'Análise de aumento dos valores de metalurgia de 2003 a 2007',
    'Metalurgia', 
    'Média Movel Simples'
)
```

#### Agora, iremos aplicar o algoritmo de Alisamento Exponencial Simples para os valores de metalurgia

```python
metalurgia = dados_agrupados['metalurgia']

fit1 = SimpleExpSmoothing(metalurgia, initialization_method="heuristic").fit(smoothing_level=0.2,optimized=False)
fcast1 = fit1.forecast(3).rename(r'$\alpha=0.2$')

fit2 = SimpleExpSmoothing(metalurgia, initialization_method="heuristic").fit(smoothing_level=0.6,optimized=False)
fcast2 = fit2.forecast(3).rename(r'$\alpha=0.6$')

fit3 = SimpleExpSmoothing(metalurgia, initialization_method="estimated").fit()
fcast3 = fit3.forecast(3).rename(r'$\alpha=%s$'%fit3.model.params['smoothing_level'])

plt.plot(dados_agrupados, marker='o', color='black')

plt.plot(fit1.fittedvalues, marker='o', color='blue')
line1, = plt.plot(fcast1, marker='o', color='blue')

plt.plot(fit2.fittedvalues, marker='o', color='red')
line2, = plt.plot(fcast2, marker='o', color='red')

plt.plot(fit3.fittedvalues, marker='o', color='green')
line3, = plt.plot(fcast3, marker='o', color='green')

plt.legend([line1, line2, line3], [fcast1.name, fcast2.name, fcast3.name])

configurar_plot('Alisamento Exponencial Simples para os dados de metalurgia', '', '')
```

### Aplicando o modelo Holt Winters nos dados de comércio

```python
dados_agrupados = dados[['data', 'comercio']].groupby('data', as_index = True).sum()

comercio = dados_agrupados['comercio']

fit1 = Holt(comercio, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2, optimized=False)
fcast1 = fit1.forecast(5).rename("Holt's linear trend")

fit2 = Holt(comercio, exponential=True, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2, optimized=False)
fcast2 = fit2.forecast(5).rename("Exponential trend")

fit3 = Holt(comercio, damped_trend=True, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2)
fcast3 = fit3.forecast(5).rename("Additive damped trend")

plt.figure(figsize=(12, 8))

plt.plot(comercio, marker='o', color='black')

plt.plot(fit1.fittedvalues, color='blue')
line1, = plt.plot(fcast1, marker='o', color='blue')

plt.plot(fit2.fittedvalues, color='red')
line2, = plt.plot(fcast2, marker='o', color='red')
plt.plot(fit3.fittedvalues, color='green')

line3, = plt.plot(fcast3, marker='o', color='green')
plt.legend([line1, line2, line3], [fcast1.name, fcast2.name, fcast3.name])

configurar_plot('Análise do modelo Holt Winters nos dados de comércio', '', '')
```

### Aplicando o modelo Holt Winters nos dados de alimento

```python
dados_agrupados = dados[['data', 'alimento']].groupby('data', as_index = True).sum()

alimento = dados_agrupados['alimento']

fit1 = Holt(alimento, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2, optimized=False)
fcast1 = fit1.forecast(5).rename("Holt's linear trend")

fit2 = Holt(alimento, exponential=True, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2, optimized=False)
fcast2 = fit2.forecast(5).rename("Exponential trend")

fit3 = Holt(alimento, damped_trend=True, initialization_method="estimated").fit(smoothing_level=0.8, smoothing_trend=0.2)
fcast3 = fit3.forecast(5).rename("Additive damped trend")

plt.figure(figsize=(12, 8))

plt.plot(alimento, marker='o', color='black')

plt.plot(fit1.fittedvalues, color='blue')
line1, = plt.plot(fcast1, marker='o', color='blue')

plt.plot(fit2.fittedvalues, color='red')
line2, = plt.plot(fcast2, marker='o', color='red')
plt.plot(fit3.fittedvalues, color='green')

line3, = plt.plot(fcast3, marker='o', color='green')
plt.legend([line1, line2, line3], [fcast1.name, fcast2.name, fcast3.name])

configurar_plot('Análise do modelo Holt Winters nos dados de alimento', '', '')
```
